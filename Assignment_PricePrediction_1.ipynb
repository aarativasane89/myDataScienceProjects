```python
# lets check for outliers using boxplot
plt.figure(figsize = (13,9))
sns.boxplot(y=data['price'])
```

```python
# Findout outliers using z-score
outliers=[]
def detect_outlier_price():
    threshold=3
    mean_1 = data['price'].mean()
    std_1 =data['price'].std()
    for y in data['price']:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    print(len(outliers))
```

```python
detect_outlier_price()
```
